```python
imgs = []
tg = TagGenerator3("tagStandard41h12")

for tag_id in [0,1,2,3,4,5,100,113,137,141,146,283]:
    im = tg.generate(tag_id, 30)
    imgs.append(im)
    
im = mosaic(imgs)
plt.imshow(im, cmap="gray")
plt.axis("off");
```
